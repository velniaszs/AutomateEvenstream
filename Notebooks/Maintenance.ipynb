```python
# --- Fabric Maintenance: Control Tables ---
# Handles the "small file problem" for high-frequency (4-min) ETL loads.

# List your 3 control tables (Format: 'LakehouseName.TableName')
control_tables = [
    "fix_aop_config", 
    "oap_load_config", 
     "dataverse_load_config"
]

for table in control_tables:
    print(f"Maintaining: {table}")
    
    # 1. Compact small files into larger ones
    # This reduces read overhead significantly for your next ETL run.
    spark.sql(f"OPTIMIZE {table}")
    
    # 2. Cleanup old, unreferenced parquet files
    # We override the 7-day safety check to clean up 4-minute garbage faster.
    spark.sql("SET spark.databricks.delta.retentionDurationCheck.enabled = false")
    
    # Keeping only 24 hours of history is usually plenty for control tables.
    spark.sql(f"VACUUM {table} RETAIN 1 HOURS")
    
    print(f"  - [OK] {table} optimized and vacuumed.")

print("Maintenance completed for all 3 tables.")
```
